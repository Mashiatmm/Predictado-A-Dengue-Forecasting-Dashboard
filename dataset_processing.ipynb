## Upsampling - Making monthly data into daily data

```python
groups = dengue_df.groupby('City')
cities = dengue_df['City'].unique().tolist()
print(cities)
```

```python
all_dfs = []
for city in cities : 
  df_city = groups.get_group(city)
  df_city = df_city.resample('D').interpolate(method='linear')
  df_city['Value'] = round(df_city['Value']/30)
  df_city['City'] =  df_city['City'].fillna(city)
  all_dfs.append(df_city)
  print(df_city)
```

```python
dengue_df_gen = pd.concat(all_dfs,axis=0)
dengue_df_gen
```

```python
dengue_df_gen.to_csv('/content/gdrive/MyDrive/Dengue2/MVH22 DATASET/dengue_generated.csv')
```

# Weather data - temperature

```python
df = pd.read_csv('/content/gdrive/MyDrive/Dengue2/MVH22 DATASET/tas_timeseries_monthly_cru_1901-2020_LKA.csv')
df.rename(columns={'Unnamed: 0' : 'year'},inplace=True)
temp_df = df
df
```

```python
temp_df = temp_df.loc[temp_df['year'] == 2019]
```

```python
new_df = pd.DataFrame()
new_df['Date']= pd.date_range(start='2019-01-01', end='2019-12-31')
new_df
```

```python
months = temp_df.columns.values.tolist()
months.remove('year')
months
```

```python
def leap_year(year):
    if year % 400 == 0:
        return True
    if year % 100 == 0:
        return False
    if year % 4 == 0:
        return True
    return False

def days_in_month(month, year):
    if month in {'Jan', 'Mar', 'May', 'Jul', 'Aug', 'Oct', 'Dec'}:
        return 31
    if month == 'Feb':
        if leap_year(year):
            return 29
        return 28
    return 30
```

```python
new_df["Temperature"] = None
tot_days=0
for month in months :
  days = days_in_month(month,2019)
  print(month,days)
  for i in range(days):
    new_df['Temperature'].iloc[i+tot_days] = round(random.uniform(temp_df[month].iloc[0]-3, temp_df[month].iloc[0]+3), 2)
  tot_days+=days
new_df
```

```python
temp_df_gen = new_df
```

# Weather data - precipitation

```python
df = pd.read_csv('/content/gdrive/MyDrive/Dengue2/MVH22 DATASET/pr_timeseries_monthly_cru_1901-2020_LKA.csv')
df.rename(columns={'Unnamed: 0' : 'year'},inplace=True)
prec_df = df
df
```

```python
new_df = pd.DataFrame()
new_df['Date']= pd.date_range(start='2019-01-01', end='2019-12-31')
new_df
```

```python
new_df["Precipitation"] = None
tot_days=0
for month in months :
  days = days_in_month(month,2019)
  print(month,days)
  for i in range(days):
    new_df['Precipitation'].iloc[i+tot_days] = round(random.uniform(prec_df[month].iloc[0]-3, prec_df[month].iloc[0]+3), 2)
  tot_days+=days
new_df
```

```python
prec_df_gen = new_df
```

```python
weather = pd.merge(temp_df_gen,prec_df_gen,how='outer',on='Date')
weather
```
